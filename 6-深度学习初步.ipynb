设直线 L 的方程为Ax+By+C=0，点 P 的坐标为（x0,y0），则点 P 到直线 L 的距离为：

此处$|Ax+By+C|=1$,因为这句：if np.max(f(label_a_x,k,b)) <= -1 and np.max(f(label_b_x,k,b)) >= 1:
$A = k,B = -1,C = b ,||w|| = (k^2+1)^\frac{1}{2}$

点到直线的距离：$$\frac{1}{||w||}$$
#### 直线上下两簇到直线最近点到直线的距离为：$$\frac{2}{||w||}$$

如果想让点到直线的距离最大，即需要$\frac{2}{||w||}$最大，从而要||w||最小，即可得到margin最大的直线,即k最小

```python
# >>> students = [('john', 'A', 15), ('jane', 'B', 12), ('dave', 'B', 10)]
# >>> sorted(students, key=lambda s: s[2])            # 按年龄排序
# [('dave', 'B', 10), ('jane', 'B', 12), ('john', 'A', 15)]


import random
plt.scatter(*zip(*label_a))
plt.scatter(*zip(*label_b))

# for k,b in k_and_b:
k,b = sorted(k_and_b,key=lambda t:abs(t[0]))[0]
print(k,b)
x = np.concatenate((label_a,label_b))
plt.plot(x,f(x,k,b))
```

#### 为了简化问题
对于一部分点，我们希望的是$w*x_i+b>1$，label=1，对于另外一部分点，我们希望的是$w*x_i+b<-1$,label=-1,可以把上述两种情况简化为：
$$label (w*x_i+b)>1$$

This margin of cloest points to this line is:$\frac{2}{||w||}$

'Minize ||w|| subject to $label (w*x_i+b)>1 for i == 1,2,3...,n$

#### 怎么获得最小的||w||?

Hingeloss
$$loss = max (0,1-label (w*x_i+b))$$

只要最近的点离直线的距离大于1，更远的点loss就是0

只需要一直求解：$$ \frac{\partial{loss}}{\partial{w}}$$

为了能够更快的找到||w||,我们要加一个regularization$$loss = max (0,1-label (w*x_i+b)) + \lambda|w|^2$$
$$loss = max (0,1-label (w*x_i+b)) + \frac{1}{2}\lambda|w|^2$$
写成$\frac{1}{2}$为了让计算机在求导时直接把2约掉

#### 多维空间，w,x,b变成向量即可

```python
radious_one = 10
radius_two = 20
```

```python
some_points = [(x,(radious_one-x**2)**0.5) for x in range (-radious_one//2,radious_one//2)]+\
[(x,-(radious_one-x**2)**0.5) for x in range (-radious_one//2,radious_one//2)]

another_points = [(x,(radius_two-x**2)**0.5) for x in range (-radius_two//2,radius_two//2)]+\
[(x,-(radius_two-x**2)**0.5) for x in range (-radius_two//2,radius_two//2)]
```

#### 在SVM中内置了很多空间变换的函数，人们把这种函数叫做核函数
+ 核函数的意义：在低维空间不能线性分割，在经过核函数的转换后，转换到高维空间，变成可以线性分割
+ 前些年比较火，最近几年用的不多，原因：没有像决策树，KNN，贝叶斯那样的可解释性，运行过程还比较复杂
+ 当数据量比较小，维度比较小的时候，用SVM 可以处理，当数据量比较大，维度比较大的时候，再增加一个维度，数据并没有变得更复杂，SVM就处理不了了

### 随机抽取数据

```python
def random_select(dataframe,y, drop_num=4): 
    columns = random.sample(list(dataframe.columns), k=len(dataframe.columns) - drop_num) #随机选取行
    Indices = np.random.choice(range(len(dataframe)),size=len(dataframe),replace=True) #随机选取列
    
    return dataframe.iloc[Indices][columns],y[Indices]
```

```python
random_select(dataframe,y, drop_num=4)[0]
```

```python
(train_x, test_x, train_y, test_y) = train_test_split(dataframe, y, test_size=0.3)

train_sample,bagging_y = random_select(train_x,train_y, drop_num=4)
```

```python
regressioner = DecisionTreeRegressor()
```

```python
regressioner.fit(train_sample, bagging_y)
```

```python
train_score = regressioner.score(train_sample, bagging_y)
test_score = regressioner.score(test_x[train_sample.columns], test_y)

print (f'training score and testing score are:{train_score} vs {test_score}')
```

```python
split_data = (train_x, test_x, train_y, test_y)
```

```python
def random_tree(train_x, test_x, train_y, test_y,drop_num=4):
    regressioner = DecisionTreeRegressor()
    regressioner.fit(train_sample, bagging_y)

    train_score = regressioner.score(train_sample, bagging_y)
    test_score = regressioner.score(test_x[train_sample.columns], test_y)

    print (f'training score and testing score are:{train_score} vs {test_score}')

    y_predict = regressioner.predict(test_x[train_sample.columns])
    
    return y_predict
```

```python
random_tree(*split_data)
```

## 回归取均值，分类投票

```python
def random_forest(train_x, test_x, train_y, test_y,drop_num=4,tree_num=4):
    forest_predict = [random_tree(train_x, test_x, train_y, test_y,drop_num) for _ in range(tree_num)]
    
    return np.mean(forest_predict,axis=0)
```

```python
forest_predict = random_forest(*split_data)
```

```python
from sklearn.metrics import r2_score
```

## Ensemble Result is Better than Single Model


```python
r2_score(test_y,forest_predict)
```

```python
forest_predict = random_forest(*split_data,tree_num=10)
```

```python
r2_score(test_y,forest_predict)
```
