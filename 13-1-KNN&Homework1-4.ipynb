```python
# 将数据变为array的形式
def filetoarray(filename):
    fr = open(filename,'r',encoding = 'utf-8')
    lines = fr.readlines()
    
    len_lines = len(lines) # 1000
    
    returnArray = np.zeros((len_lines,3))

    # 标签
    classLabel = []
    index = 0
    
    for line in lines:
        line = line.strip().split('\t')
        # 写入index行
        returnArray[index,:] = line[0:3]
        
        if line[-1] == 'didntLike':
            classLabel.append(1)
        elif line[-1] == 'smallDoses':
            classLabel.append(2)
        elif line[-1] == 'largeDoses':
            classLabel.append(3)
        index =+ 1
        
    return returnArray,classLabel
```

```python
# dataSet 特征
def autoNorm(dataSet):
    # 获取数据的最小值 Min
    minValue = dataSet.min()
    # 获取数据的最大值 Max
    maxValue = dataSet.max()
    
    # 最大最小值的范围，即分母
    ranges = maxValue - minValue
    
    # 获得行列为dataSet的shape的0 array
    normDataSet = np.zeros(np.shape(dataSet))
    # 特征有多少行
    m = dataSet.shape[0]
    
    normDataSet = dataSet - np.tile(minValue,(m,1)) # 把最小值复制m行，分子
    normDataSet = normDataSet/np.tile(ranges,(m,1)) # ranges是一个值，要复制这么多行，才可以进行运算
    
    return normDataSet,minValue,ranges
```

```python
# A:(x1,y1,z1),B:(x2,y2,z2),两点间距离：((x2-x1)**2 + (y2-y1)**2 + (z2-z1)**2)**0.5
# 定义一个分类器
"""
input:
    inX:(feature1,feature2,feature3) 后来的测试数据
    dataSet:feature_array
    labels:分类标签
    k = 选择距离最小的k个点
"""
def classify(inX,dataSet,labels,k):
    dataSetsize = dataSet.shape[0]
    
    # sqDistances=(x2-x1)**2 + (y2-y1)**2 + (z2-z1)**2
    # 相减
    diffMat = np.tile(inX,(dataSetsize,1)) - dataSet
    # 二维特征相减后平方
    sqDiffMat = diffMat ** 2
    # 相减后相加,axis=0列相加，axis=1行相加
    sqDistances = sqDiffMat.sum(axis=1)
    
    # sqDistances**0.5
    distance = sqDistances**0.5
    
    # 排序
    sortDistance = distance.argsort()
    
    # 定义一个记录类别次数测字典
    classCount = {}
    # 选取k个点
    for i in range(k):
        # 取出前k个元素的类别
        voteLabel = labels[sortDistance[i]]
        # 计算类别次数
        classCount[voteLabel] = classCount.get(voteLabel,0)+1
        
    sortedClassCount = sorted(classCount.items(),key=operator.itemgetter(1),reverse=True)
    print(sortedClassCount)
    
    return sortedClassCount[0][0]
```

```python
# 测试函数
def datingClassTest():
    # 打开文件
    filename = './datingTestSet.txt'
    dataArray,labels = filetoarray(filename)
    # 数据归一化
    normArray,minVals,ranges = autoNorm(dataArray)
    # 获取normArray行数
    m = normArray.shape[0]
    
    # 取10%的数据作为测试集
    hoRatio = 0.1
    numTestVacs = int(m*hoRatio)
    
    # 分类错误计数
    errorCount = 0.0
    for i in range(numTestVacs):
        # 前numTestVacs个数据作为测试集，后m-numTestVacs个数据作为训练
        # 90%--trainSet,10%--testSet
        """
        input:
            inX:(feature1,feature2,feature3) 后来的测试数据
            dataSet:feature_array
            labels:分类标签
            k = 选择距离最小的k个点
        """
        classifierResult = classify(normArray[i],normArray[numTestVacs:m,:],labels[numTestVacs:m],4)
        print('分类结果：%s\t真实类别：%d'%(classifierResult,labels[i]))
        if classifierResult != labels[i]:
            errorCount +=1
            
    print('错误率：{}%'.format(errorCount/float(numTestVacs)*100))
        
```

```python
# 现在来了一位男士，测试一下是不是海伦的菜
def classifyPerson():
    # 输出结果
    resultList = ['讨厌','有些喜欢','喜欢']
    # inX -- 三个特征
    percenTats = float(input("玩游戏时间占比： "))
    ffMiles = float(input("飞机里程数： "))
    iceCream = float(input("每周消费冰淇淋数： "))
    
    # 打开文件
    filename = './datingTestSet.txt'
    dataArray,labels = filetoarray(filename)
    # 数据归一化 normArray是归一化之后的测试集
    normArray,minVals,ranges = autoNorm(dataArray)
    
    # 三个float类型的特征转换成array
    inX = np.array([ffMiles,percenTats,iceCream])
    
    # inX归一化
    norminArr = (inX - minVals) / ranges
    
    # 分类
    classfierResult = classify(norminArr,normArray,labels, 3)
    
    # 打印结果
    print('海伦，你可能{}这个人'.format(resultList[classfierResult-1]))
```

# 2 数据预处理

## 2.1 为向量化表示进行前处理
进行自然语言处理时，必须将单词转换为机器学习算法可以利用的向量。如果目标是对文本数据进行机器学习建模，例如电影评论或推文或其他任何内容，则需要将文本数据转换为数字。此过程称为“嵌入”或“向量化”。  
进行向量化时，请务必记住，它不仅仅是将单个单词变成单个数字。单词可以转换为数字，整个文档就可以转换为向量。向量的维度往往不止一个，而且对于文本数据，向量通常是高维的。这是因为特征数据的每个维度将对应一个单词，而我们所处理的文档通常包含数千个单词。

## 2.2 TF-IDF
在信息检索中，tf–idf 或 TFIDF（term frequency–inverse document frequency）是一种数值统计，旨在反映单词对语料库中文档的重要性。在信息检索，文本挖掘和用户建模的搜索中，它通常用作加权因子。 tf-idf 值与单词在文档中出现的次数成正比，同时被单词在语料库中的出现频率所抵消，这有助于调整某些单词通常会更频繁出现的事实。 如今，tf-idf是最流行的术语加权方案之一。在数字图书馆领域，有83％的基于文本的推荐系统使用tf-idf。

搜索引擎经常使用tf–idf加权方案的变体作为在给定用户查询时对文档相关性进行评分和排名的主要工具。tf–idf可成功用于各种领域的停用词过滤，包括文本摘要和分类。

排名函数中最简单的是通过将每个查询词的tf–idf相加得出，许多更复杂的排名函数是此简单模型的变体。

> TF-IDF（Term Frequency-inverse Document Frequency）是一种针对关键词的统计分析方法，用于评估一个词对一个文件集或者一个语料库的重要程度。一个词的重要程度跟它在文章中出现的次数成正比，跟它在语料库出现的次数成反比。

>  一个词在语料库中的 TF-IDF = TF * IDF

>> TF = 当前文档中的词频 / 当前文档中所有词的数量  
IDF = log{ 所有文档的总数 / （包含该词有多少个文档数量+1）}

#### 当某个词不存在与文档中时，如果不加1，IDF log的分母变为0，+1是进行一个平滑

举个例子： 数据集中一共有三个文档（举例子的小文档）

【我 可 太喜欢 姚明 了】，【谁 的 老婆 是 叶莉 啊】，【叶莉 有 一个 女儿 和 一个 儿子】  

5，6，7个词

【我 可 太喜欢 姚明 了】 中的 “姚明” TF-IDF = (1/5) * log(3/(1+1))

# 3 K-Means 聚类

## 3.1 使用手肘法选择聚类簇的数量
随着聚类数k的增大,样本划分会更加的精细,每个簇的聚合程度会逐渐提高,那么误差平方和SSE自然会逐渐变小,并且当k小于真实的簇类数时,由于k的增大会大幅增加每个簇的聚合程度,因此SSE的下降幅度会很大,而当k到达真实聚类数时,再增加k所得到的聚合程度回报会迅速变小,所以SSE的下降幅度会骤减,然后随着k值的继续增大而趋于平缓,也就是说SSE和k的关系类似于手肘的形状,而这个肘部对应的k值就是数据的真实聚类数.因此这种方法被称为手肘法.
